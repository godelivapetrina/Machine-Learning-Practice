```python
# Use K-NN Regressor
# Find neighbors by using all the numeric columns in the dataset to generate similarity scores.
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
```

```python
# The columns that we will be making predictions with.
x_columns = ['age', 'g', 'gs', 'mp', 'fg', 'fga', 'fg.', 'x3p','x3pa','x3p.','x2p', 'x2pa','x2p.', 'efg.', 'ft', 'fta', 'ft.', 'orb', 'drb', 'trb', 'ast', 'stl', 'blk', 'tov', 'pf']
```

```python
# The column that we are predicting (target variable)
y_column = ["pts"]
```

```python
#Pick relevant columns
X = nba[x_columns].values
Y = nba[y_column].values
```

```python
#split data set into train and test set
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2) 
```

```python
# Create the knn model
# Use 10 neighbors, inverse-distance as weights
# Algo can use {‘auto’, ‘ball_tree’, ‘kd_tree’, ‘brute’}
# p=2 means Euclidean distance
knn = KNeighborsRegressor(n_neighbors=10,weights='distance',algorithm='auto',p=2)
```

```python
# Fit the model on the training data.
knn.fit(X_train, Y_train)
```

```python
# Make point predictions on the test set using the fit model.
predictions = knn.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
#Compute error using MSE
mean_squared_error(Y_test,predictions)
```
