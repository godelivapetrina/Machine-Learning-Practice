```python
import pandas as pd
import quandl
```

```python
df = quandl.get("WIKI/GOOGL")
```

```python
df.head()
```

```python
# choosing relevant columns
df = df[['Adj. Open',  'Adj. High',  'Adj. Low',  'Adj. Close', 'Adj. Volume']]
```

```python
#Adding new columns
df['HL_PCT'] = (df['Adj. High'] - df['Adj. Low']) / df['Adj. Low'] * 100.0
```

```python
df['PCT_change'] = (df['Adj. Close'] - df['Adj. Open']) / df['Adj. Open'] * 100.0
```

```python
# define new data
df = df[['Adj. Close', 'HL_PCT', 'PCT_change', 'Adj. Volume']]
print(df.head())
```

```python
import math
import numpy as np
from sklearn import preprocessing
```

```python
forecast_col = 'Adj. Close'
#handle missing values
df.fillna(value=-99999, inplace=True)
#forecast out 1% of the entire length of the dataset.
forecast_out = int(math.ceil(0.01 * len(df)))
```

```python
df.shape
```

```python
df
```

```python
df['label'] = df[forecast_col].shift(-forecast_out)
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
#Scikit-learn actually fundamentally requires numpy arrays. 
# Pandas dataframes can be easily converted to NumPy arrays
X = np.array(df.drop(labels='label',axis=1))
```

```python
X
```

```python
Y = np.array(df['label'])
```

```python
Y
```

```python
# Do some data preprocessing
# Generally, you want your features in machine learning to be in a range of -1 to 1. 
# This may do nothing, but it usually speeds up processing and can also help with accuracy.
X = preprocessing.scale(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
# create training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2) 
```

```python
from sklearn.svm import SVR
import numpy as np
```

```python
from sklearn.model_selection import GridSearchCV
# use gridsearchcv for selecting parameters
# grid search C, gamma of rbf kernel
# For an initial search, a logarithmic grid with basis
# 10 is often helpful. Using a basis of 2, a finer
# tuning can be achieved but at a much higher cost.

C_range = np.logspace(-1, 5, 5)
gamma_range = np.logspace(-5, 3, 5)
param_grid = dict(gamma=gamma_range, C=C_range)

grid = GridSearchCV(estimator = SVR(kernel='rbf'), param_grid = param_grid,cv=3)
grid_result = grid.fit(X_train, Y_train)
```

```python
# summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
# Create model - see documentation for parameter details
clf = SVR(C=100000,kernel='rbf',gamma=0.001)
```

```python
# Fit data
clf.fit(X_train, Y_train)
```

```python
# Get parameters for this estimator.
clf.get_params()
```

```python
# Test on testset
prediction = clf.predict(X_test)
```

```python
#Returns the coefficient of determination R^2 of the prediction.
Rsquared = clf.score(X_test,Y_test)
```
